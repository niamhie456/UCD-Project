```python
# Table demonstating Manufacturer by model count and % count of the data.
Veh_brandcount = vehicledata_cleaned.groupby("Manufacturer").agg({'model':'count'})
sorted_Veh_brandcount = Veh_brandcount.sort_values(by='model', ascending=False)
sorted_Veh_brandcount["percent_total"] = vehicledata_cleaned['Manufacturer'].value_counts(normalize=True) * 100
print(sorted_Veh_brandcount)
```

```python
pricelux = vehicledata_cleaned.loc[((vehicledata_cleaned["year"] > 2010) & (vehicledata_cleaned["vehicle_type"] == "Luxury"))]
priceyear = pricelux[["price","year"]].groupby('year').mean()
priceyear.sort_values(by=['price'], inplace=True, ascending=False)
pricestand = vehicledata_cleaned.loc[((vehicledata_cleaned["year"] > 2010) & (vehicledata_cleaned["vehicle_type"] == "Standard"))]
priceyear2 = pricestand[["price","year"]].groupby('year').mean()
priceyear2.sort_values(by=['price'], inplace=True, ascending=False)

plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.title("Average Spend by Year Luxery Vehicle")
sns.heatmap(data=priceyear)
plt.xlabel("price")

plt.subplot(2,2,2)
plt.title("Average Spend by Year Standard Vehicle")
sns.heatmap(data=priceyear2)
plt.xlabel("price")
```

```python
# I really expected year to be the most deciding factor in price of the vehicle, but as the scatterplots and heatmap show, the yearer the car, doesn't actually mean the highest price.  Luxury vehicle data starts in 2011 and ends in 2017, standard starts in 2016 and ends in 2020.
```
