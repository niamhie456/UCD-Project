```python
sns.countplot(data=vehicledata_cleaned,  y='Manufacturer').set(title='Manufacturer Count')
plt.savefig('Manucount.png', dpi=1200, format='png')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.displot(data=vehicledata_cleaned, x="price", y='Manufacturer', hue= "vehicle_type").set(title='Manufacturer Price')
plt.axvline(x=vehicledata_cleaned.price.mean(), color='red')
plt.savefig('Manuprice.png', dpi=1200, format='png')
plt.show()
# There are clearly 3 Manufactueres that command a higher retail price, Audi, BMW and Mercedes. As mentioned before these are luxury brands, no surprises here. We can see theat the standard vehicles prices are concentrated under €25,000
```

```python
# we will just have a look at the average price for per Manurfacturer.
Veh_standav = vehicledata_cleaned.groupby("Manufacturer").agg({'price':'mean'})
sorted_Veh_standav = Veh_standav.sort_values(by='price', ascending=False)
round_Veh_standav = np.round_(sorted_Veh_standav)
print(round_Veh_standav)
```

```python
#Odometer
sns.scatterplot(data=vehicledata_cleaned, x="price", y='mileage', hue="vehicle_type").set(title='Odometer with Price')
plt.axvline(x=vehicledata_cleaned.price.mean(), color='red')
plt.savefig('Odometeruprice.png', dpi=1200, format='png')
plt.show()
#Scatterplot shows a clear correlation between price and milage, high mileage, lower price. The diagram shows luxury vehicles with very low mileage, demanding a very high price. We can also see a difference in price and mileage for luxury/standard vehicles above £20,000 for example a standard vehicle with 50k miles will fetch approx £30,000, where a luxury vehicle with 75k miles will have a price of £30,000, thats 2 years more driving! (10k miles average per year in UK. https://www.carbuyer.co.uk/tips-and-advice/303702/used-car-mileage-uk-average-and-how-much-is-too-much).  this scatter plot also shows the majority of the data is below 120k mileage and below 60k in price.
```

```python
Year_Price = sns.histplot(data=vehicledata_cleaned,  x='year', y="price", hue="vehicle_type", bins=20, legend=True).set(title='Year and Price')
plt.savefig('yearprice.png', dpi=1200, format='png')
plt.show()
#I tried to move the legend with sns.move_legend(Year_Price, "center left") and plt.legend(loc='upper left').
```

```python
pricelux = vehicledata_cleaned.loc[((vehicledata_cleaned["year"] > 2010) & (vehicledata_cleaned["vehicle_type"] == "Luxury"))]
priceyear = pricelux[["price","year"]].groupby('year').mean()
priceyear.sort_values(by=['price'], inplace=True, ascending=False)
pricestand = vehicledata_cleaned.loc[((vehicledata_cleaned["year"] > 2010) & (vehicledata_cleaned["vehicle_type"] == "Standard"))]
priceyear2 = pricestand[["price","year"]].groupby('year').mean()
priceyear2.sort_values(by=['price'], inplace=True, ascending=False)

plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.title("Average Spend by Year Luxery Vehicle")
sns.heatmap(data=priceyear)
plt.xlabel("price")

plt.subplot(2,2,2)
plt.title("Average Spend by Year Standard Vehicle")
sns.heatmap(data=priceyear2)
plt.xlabel("price")
plt.savefig('heatmap.png', dpi=1200, format='png')
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(4,2,1)
sns.scatterplot(x='year',y ='price',hue="fuelType",data=vehicledata_cleaned).set(title='Year and Price')
plt.subplot(4,2,2)
sns.scatterplot(x='year',y ='price',hue="transmission",data=vehicledata_cleaned).set(title='Year and Price')
plt.subplot(4,2,3)
sns.scatterplot(x='year',y ='price',hue="vehicle_type",data=vehicledata_cleaned).set(title='Year and Price')
plt.subplot(4,2,4)
sns.scatterplot(x='year',y ='mileage',hue="fuelType",data=vehicledata_cleaned).set(title='Year and mileage')
plt.savefig('yearprice1.png', dpi=1200, format='png')
plt.show()
```
